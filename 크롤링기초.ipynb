### enumerate란?

```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('./chromedriver') #chrome창 자체를 열어줌
```

```python
#페이지가 모두 로드될 때까지 3초 기다림
driver.implicitly_wait(3)
```

```python
#셀레니움을 통해 실제 크롬창에서 접속한 것과 동일하게 해당 URL 내용을 가져옴
driver.get(url)
```

```python
#소스를 읽어 Beautiful Soup을 이용해 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
#이전에 requests와 BeautifulSoup을 이용했던 코드와 동일하게 처리
rank_list = soup.select(".ranking_list .item_title")
rank_result_list=[]
for (index,rank) in enumerate(rank_list):
    rank_result_list += [{"rank": index+1, "value": rank.text}]
```

```python
print(rank_result_list)
print("-"*100)
print(f'실시간 급상승 검색어 {rank_result_list[0]["rank"]}위는 "{rank_result_list[0]["value"]}"입니다.')
```

### Selenium과 Request로 크롤링 하는 것의 차이

웹서버는 우리가 사람(브라우저)인 것처럼 판단하고 있습니다. 만약 단순히 파이썬 requests 라이브러리만을 이용했다면 서버는 우리가 사람이 아닌 것을 알 수 있습니다.
